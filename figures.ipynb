```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from platform import python_version
print(python_version())
print(matplotlib.__version__)
print(np.__version__)
```

```python
plt.rcParams["figure.figsize"] = (5,3)

labels = ['Penetration\nTesting', 'Security-\nby-Design', 'Security\nStandards']
no = [0, 0, 2]
no_planning = [1,2,3]
yes = [13, 12, 9]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()

no_rects = ax.bar(x - width, no, width, label='No', color='firebrick', edgecolor='black', linewidth=0.5)
no_planning_rects = ax.bar(x, no_planning, width, label='No, but planning', color = 'darkorange', edgecolor='black', linewidth=0.5)
yes_rects = ax.bar(x + width, yes, width, label='Yes', color='forestgreen', edgecolor='black', linewidth=0.5)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Absolute Frequency', fontsize=12)
# ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels, fontsize=12)
ax.legend(loc='upper right', fontsize=8)
ax.grid(alpha=0.5)
#ax.bar_label(no, padding=3)
#ax.bar_label(no_planning, padding=3)
ax.set_ylim(0,14)
ax.set_title("$[\mathbb{C}]~Do~you~adopt~these~cybersecurity~strategies~in~your~company?$", fontsize=10)

fig.tight_layout()
plt.savefig("figures/strategy_new.pdf")
plt.show()
```

```python
pa = [8,5,10,10]
comp = [8,5,9,7]
aca = [2,8,3]

data = {'Public\nAuthorities': pa, 'Other\nCompanies': comp, 'Academia':aca}
medianprops = dict(linestyle='-', linewidth=1, color='black')
# Random test data

labels = ['x1', 'x2', 'x3']

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(3.5, 3))

# rectangular box plot
bplot1 = ax1.boxplot(data.values(), widths=0.8, medianprops=medianprops,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks

# fill with colors
colors = ['yellowgreen', 'lightblue', 'lightgrey']
for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)

        
# adding horizontal grid lines
for ax in [ax1]:
    ax.yaxis.grid(True)
    ax.set_ylabel('Willingness to share [1-10]', fontsize=10)
    ax.set_xticklabels(data.keys())
    ax.set_ylim(0.5,10.5)
ax.set_title("$[\mathbb{A}]~How~much~are~\mathbb{C}~willing~to~share~data~with...$", fontsize=10)
plt.tight_layout()
plt.savefig("figures/sharing_new.pdf")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (6,4)
fig, ax = plt.subplots()

# Example data
attacks = ('FDI', 'Data Breach', 'Spoofing', 'MitM', 'APT', 'Malware', 'DoS')
y_pos = np.arange(len(attacks))
performance = [0.75, 0, 0, 0, 1, 0.75, 0.5]

ax.barh(y_pos, performance, align='center', edgecolor='black', linewidth=0.5, color='yellowgreen')
ax.set_yticks(y_pos, labels=attacks)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Relative Frequency', fontsize=12)
ax.set_title("$[\mathbb{A}]~What~are~the~most~dangerous~threats~to~the~SG?$", fontsize=12)
ax.grid(alpha=0.5)
plt.tight_layout()
plt.savefig("figures/attacks_authorities_new.pdf")
plt.show()
```
